```python
import numpy as np

np.random.seed(0)

# Initialize an empty list to store the data for each cluster
clusters = []
labels = []

# Define the number of samples for each cluster
cluster_sizes = np.linspace(10, 2000, 25).astype(int)

# For each cluster
for i in range(25):
    # Generate a random mean value between -10 and 10 for each dimension
    mean = np.random.uniform(-10, 10, 20)
    
    # Generate a random covariance matrix
    cov = np.random.rand(20, 20)
    cov = np.dot(cov, cov.transpose())  # to make it positive semi-definite
    
    # Generate data for this cluster
    cluster = np.random.multivariate_normal(mean, cov, cluster_sizes[i])
    
    # Append this cluster to the list of clusters
    clusters.append(cluster)
    
    # Generate labels for this cluster
    cluster_labels = np.full(cluster_sizes[i], i)
    labels.append(cluster_labels)

# Concatenate the clusters to get the final dataset
X = np.concatenate(clusters, axis=0)

# Concatenate the labels to get the final target array
y = np.concatenate(labels)
```

```python
start = time.time()
km = KMeans(n_clusters=25, random_state = 0, n_init = 'auto').fit(X)
end = time.time()
print('Kmeans puro' , f'Tempo: {round(end-start,3)}', sep = '\n')
print(f'Precisão: {precision_score(y, mapeamento_clusters(y, 25, X,(km.labels_, km.cluster_centers_)), average="weighted")}',end = '\n\n')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize seed for reproducibility
np.random.seed(0)

# Define mean and covariance for each cluster
cluster_1 = {"mean": (1, 1), "cov": [[1, 0], [0, 1]]}
cluster_2 = {"mean": (6, 6), "cov": [[2, 1], [1, 2]]}
cluster_3 = {"mean": (1, 6), "cov": [[1, 0], [0, 1]]}
clusters = [cluster_1, cluster_2, cluster_3]

# Number of points per cluster
num_points = 100

# Generate the clusters
data = np.concatenate([
    np.random.multivariate_normal(c["mean"], c["cov"], num_points)
    for c in clusters
])

# Plot the points
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
from scipy.stats import gaussian_kde

# Calculate the kernel density estimation
kde = gaussian_kde(data.T)

# Create a grid over which to evaluate the kde
x = np.linspace(-2, 8, 1000)
y = np.linspace(-2, 8, 1000)
X, Y = np.meshgrid(x, y)
positions = np.vstack([X.ravel(), Y.ravel()])

# Evaluate the kde over the grid
Z = np.reshape(kde(positions).T, X.shape)

# Plot the kde
fig, ax = plt.subplots()
ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r,
          extent=[-2, 8, -2, 8])
ax.plot(data[:, 0], data[:, 1], 'k.', markersize=2)
ax.set_xlim([-2, 8])
ax.set_ylim([-2, 8])
plt.show()
```
